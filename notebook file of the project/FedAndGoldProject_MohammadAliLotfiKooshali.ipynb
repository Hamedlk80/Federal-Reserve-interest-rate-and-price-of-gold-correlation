## Total Analysis of relationship between FFR and CLOSE price of GOLD

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.tsa.seasonal import seasonal_decompose

data = pd.read_csv('GOLDFFR-cleaned.csv', parse_dates=['Date'], index_col='Date')

print(data.head())
```

```python
# Ensure there are no missing values
data = data.dropna()

# Resample the data to monthly frequency if necessary
data_monthly = data.resample('M').last()

# Ensure the index has a frequency set
data_monthly.index.freq = 'M'

print(data_monthly.head())
```

```python
# Plotting Gold Prices and FFR over time
plt.figure(figsize=(14, 7))

# Plot Gold Prices
plt.subplot(2, 1, 1)
plt.plot(data_monthly.index, data_monthly['Close'], color='gold', label='Gold Prices')
plt.title('Gold Prices (2001-2024)')
plt.ylabel('Price (USD)')
plt.legend()

# Plot FFR
plt.subplot(2, 1, 2)
plt.plot(data_monthly.index, data_monthly['DFF'], color='blue', label='Federal Funds Rate')
plt.title('Federal Funds Rate (2001-2024)')
plt.xlabel('Date')
plt.ylabel('Rate (%)')
plt.legend()

plt.tight_layout()
plt.show()

# Scatter plot with trend line
plt.figure(figsize=(10, 6))
sns.regplot(x='DFF', y='Close', data=data_monthly, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.title('Relationship Between Gold Prices and Federal Funds Rate')
plt.xlabel('Federal Funds Rate (%)')
plt.ylabel('Gold Prices (USD)')
plt.show()
```

```python
# Calculate correlation
correlation = data_monthly[['Close', 'DFF']].corr()
print("Correlation Matrix:\n", correlation)

# Prepare the data for regression
X = data_monthly['DFF']
y = data_monthly['Close']

# Add a constant to the independent variable
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print(model.summary())
```

```python
# Rolling correlation
rolling_corr = data_monthly['Close'].rolling(window=12).corr(data_monthly['DFF'])

plt.figure(figsize=(14, 7))
plt.plot(data_monthly.index, rolling_corr, label='Rolling Correlation (12 months)')
plt.title('Rolling Correlation between Gold Prices and Federal Funds Rate')
plt.xlabel('Date')
plt.ylabel('Correlation')
plt.legend()
plt.show()
```

```python
# Ensure the index has a frequency set
data_monthly.index.freq = 'M'

# Time series decomposition
decomposition = seasonal_decompose(data_monthly['Close'], model='additive')
fig = decomposition.plot()
plt.show()
```
